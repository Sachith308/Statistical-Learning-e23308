```python
from data_inspector import DataInspector

print("--- Step 1: Initialization & Upload ---")
# Upload the Titanic dataset CSV when prompted
inspector = DataInspector.upload_data()

if inspector:
    print("\n--- Step 2: Structural Analysis ---")
    inspector.data_summary()

    print("\n--- Step 3: Cleaning ---")
    inspector.remove_duplicates()
    inspector.handle_missing_values(strategy='median')

    # Check for Fare outliers in Titanic
    if 'Fare' in inspector.df.columns:
        inspector.handle_outliers(column='Fare', action='flag')

    print("\n--- Step 4: Feature Engineering ---")
    unified_data = inspector.get_unified_dataframe()
    print("Merged Numeric and Encoded Categorical Data:")
    display(unified_data.head())

    print("\n--- Step 5: Advanced Visualizations ---")
    if 'Age' in inspector.df.columns:
        inspector.visualize_univariate('Age')

    if 'Age' in inspector.df.columns and 'Survived' in inspector.df.columns:
        # Cast to string to force categorical plotting logic
        inspector.df['Survived'] = inspector.df['Survived'].astype(str)
        inspector.visualize_relationship('Survived', 'Age')

    print("\n--- Step 6: Deep Statistical Insights ---")
    # This will take a moment as it calculates Pearson, Cramer's V, and ANOVA
    inspector.plot_all_associations_heatmap()
```
