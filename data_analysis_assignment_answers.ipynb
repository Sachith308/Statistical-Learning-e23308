##Part A

1. Derivation of the Bias in the MLE Covariance Matrix


$$\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T$$


where $\widehat{\boldsymbol{\mu}}_n = \frac{1}{n}\sum_{i=1}^n \mathbf{X}_i$ is the empirical sample mean.

To determine if this estimator is biased, expand the deviations around the true, hidden population mean $\boldsymbol{\mu}$:


$$\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n = (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})$$

Substituting this back into the sum inside the MLE formula gives:


$$\sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T = \sum_{i=1}^n \left[ (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right] \left[ (\mathbf{X}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right]^T$$

Expanding the product matrix term-by-term yields:


$$= \sum_{i=1}^n (\mathbf{x}_i - \boldsymbol{\mu})(\mathbf{x}_i - \boldsymbol{\mu})^T - \sum_{i=1}^n (\mathbf{x}_i - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T - \sum_{i=1}^n (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\mathbf{x}_i - \boldsymbol{\mu})^T + \sum_{i=1}^n (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T$$

 $\sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu}) = n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})$.  Using this simplification, the two middle cross-product terms contract:


$$= \sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T - n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T - n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T + n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T$$

$$= \sum_{i=1}^n (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T - n(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T$$

Apply the expectation operator $\mathbb{E}[\cdot]$ to both sides:


$$\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{1}{n} \sum_{i=1}^n \mathbb{E}[(\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T] - \mathbb{E}[(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T]$$

By definition of the structural population moments:

* $\mathbb{E}[(\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_i - \boldsymbol{\mu})^T] = \boldsymbol{\Sigma}$
* $\mathbb{E}[(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T] = \text{Var}(\widehat{\boldsymbol{\mu}}_n) = \frac{1}{n}\boldsymbol{\Sigma}$

Substituting these back into our expectation equation gives:


$$\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{1}{n} (n\boldsymbol{\Sigma}) - \frac{1}{n}\boldsymbol{\Sigma} = \boldsymbol{\Sigma} - \frac{1}{n}\boldsymbol{\Sigma} = \left(\frac{n-1}{n}\right)\boldsymbol{\Sigma}$$

Because $\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] \neq \boldsymbol{\Sigma}$, the estimator is systematically downward biased. This occurs because calculating variations against the *sample* mean rather than the *true* population mean artificially underrepresents the real spatial scattering of the system.

**Bessel's Correction Elimination of Bias**

To yield an unbiased estimator ($\mathbf{S}$),scale the maximum likelihood matrix by the reciprocal of the bias factor ($\frac{n}{n-1}$):


$$\mathbf{S} = \frac{n}{n-1} \widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{n}{n-1} \left( \frac{1}{n} \sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T \right) = \frac{1}{n-1} \sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T$$

Evaluating the expectation confirms its lack of bias:


$$\mathbb{E}[\mathbf{S}] = \frac{n}{n-1} \mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{n}{n-1} \left(\frac{n-1}{n}\right)\boldsymbol{\Sigma} = \boldsymbol{\Sigma}$$

**Operational Error Diagnostics**

* **Type I Error ($\alpha$):** A **False Alarm**. This happens when the diagnostic system concludes that the structure is damaged or unstable (rejects $H_0$), when in reality the physical system remains completely nominal and healthy.
* **Type II Error ($\beta$):** A **Missed Detection**. This occurs when a structural shift or localized crack has compromised the asset (rejects $H_1$), but the diagnostic system fails to alert the team (fails to reject $H_0$).

When $\alpha$ is reduced dramatically to eliminate false alarms, the critical value boundary of the test statistic is pushed deep into the extreme tails of the baseline distribution.

1. **Impact on Statistical Power ($1-\beta$):** Because the threshold for detecting anomalies becomes incredibly stringent, minor or emerging structural shifts will fail to cross the new partition line. Therefore, $\beta$ (missed detections) will sharply escalate, causing the **statistical power ($1-\beta$) of the model to collapse**.
2. **Impact on Confidence Ellipsoids:** The confidence level of the multi-dimensional boundary scales as $1-\alpha = 0.9999$. This means the **geometric volume of the healthy operational ellipsoid expands significantly**. The boundary swells to encompass almost all variation, making the system highly insensitive to genuine baseline degradation.

##Part B

Let $\{X_n\}$ and $\{Y_n\}$ be sequences of random variables (or vectors). If $X_n$ converges in distribution to a random variable $X$ ($X_n \xrightarrow{d} X$), and $Y_n$ converges in probability to a constant scalar or matrix $c$ ($Y_n \xrightarrow{p} c$), then:

* $X_n + Y_n \xrightarrow{d} X + c$
* $X_n Y_n \xrightarrow{d} cX$
* $X_n / Y_n \xrightarrow{d} X/c$ (provided $c \neq 0$)

**Operational Distribution Justification**

The Lindeberg-Lévy Multivariate CLT asserts that:


$$\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \xrightarrow{d} \mathscr{N}(\mathbf{0}, \boldsymbol{\Sigma})$$

Multiplying by the inverse matrix square root of the population covariance normalizes the space:


$$\boldsymbol{\Sigma}^{-1/2} \left[ \sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right] \xrightarrow{d} \mathscr{N}(\mathbf{0}, \mathbf{I})$$

By the Weak Law of Large Numbers (WLLN), the unbiased empirical sample covariance matrix $\mathbf{S}$ is a continuous, consistent estimator of the population matrix $\boldsymbol{\Sigma}$. Therefore, as $n \to \infty$, $\mathbf{S}$ converges in probability to the true constant profile:


$$\mathbf{S} \xrightarrow{p} \boldsymbol{\Sigma} \implies \mathbf{S}^{-1/2} \xrightarrow{p} \boldsymbol{\Sigma}^{-1/2}$$

Applying **Slutsky's Theorem**, we multiply the sequence converging in distribution by the sequence converging in probability to a constant matrix:


$$\mathbf{S}^{-1/2} \left[ \sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right] \xrightarrow{d} \boldsymbol{\Sigma}^{-1/2} \cdot \mathscr{N}(\mathbf{0}, \boldsymbol{\Sigma}) = \mathscr{N}(\mathbf{0}, \mathbf{I})$$

Because the target boundary converges asymptotically to a standard Gaussian space, we scale the localized coordinate parameters back to the empirical matrix metric:


$$\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \sim \mathscr{N}(\mathbf{0}, \mathbf{S}) \implies \widehat{\boldsymbol{\mu}}_n \sim \mathscr{N}\left(\boldsymbol{\mu}, \frac{1}{n}\mathbf{S}\right)$$

This mathematically validates substituting $\mathbf{S}$ for $\boldsymbol{\Sigma}$ without corrupting the stability of the asymptotic diagnostic envelope.

## **Part C: Numerical Verification**

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# --- DO NOT MODIFY: Synthetic Strain Sensor Data Generator ---
np.random.seed(42)
n_samples = 5000
n_features = 3

# Generating base stationary multivariate normal noise
base_data = np.random.multivariate_normal(
    mean=[0.5, -0.2, 1.1],
    cov=[[0.09, 0.02, 0.01],
         [0.02, 0.06, 0.03],
         [0.01, 0.03, 0.05]],
    size=n_samples
)

# Injecting a subtle, hidden structural drift in the final 1,000 snapshots
base_data[4000:, 0] += 0.015  # Sensor 1 drift
base_data[4000:, 2] -= 0.010  # Sensor 3 drift

df_strain = pd.DataFrame(base_data, columns=['Strain_Ch1', 'Strain_Ch2', 'Strain_Ch3'])
# ----------------------------------------------------------------

def verify_first_moment_homogeneity(df: pd.DataFrame, g_chunks: int = 5) -> dict:
    """
    Partitions the dataset into g chunks and evaluates first-moment homogeneity
    via Wilks' Lambda and Bartlett's Chi-Square asymptotic transformation.
    """
    matrix_data = df.to_numpy()
    total_n, m = matrix_data.shape

    # Global spatial center
    mu_global = np.mean(matrix_data, axis=0)

    # Isolate into non-overlapping temporal chunks
    time_blocks = np.array_split(matrix_data, g_chunks)

    # Initialize dispersion matrices
    W_matrix = np.zeros((m, m))
    B_matrix = np.zeros((m, m))

    for block in time_blocks:
        block_n = block.shape[0]
        mu_chunk = np.mean(block, axis=0)

        # Accumulate Within-Chunk Sum of Squares (Local variations)
        mean_centered_block = block - mu_chunk
        W_matrix += mean_centered_block.T @ mean_centered_block

        # Accumulate Between-Chunk Sum of Squares (Global drifts)
        chunk_displacement = (mu_chunk - mu_global).reshape(m, 1)
        B_matrix += block_n * (chunk_displacement @ chunk_displacement.T)

    # Calculate Wilks' Lambda statistic
    lambda_wilks = np.linalg.det(W_matrix) / np.linalg.det(W_matrix + B_matrix)

    # Apply Bartlett's Chi-Square asymptotic transformation
    chi2_stat = -1.0 * (total_n - 1 - ((m + g_chunks) / 2)) * np.log(lambda_wilks)

    # Degrees of freedom constraint: m * (g - 1)
    dof = m * (g_chunks - 1)

    # Calculate statistical significance
    p_val = stats.chi2.sf(chi2_stat, dof)

    # Evaluate decision boundary against alpha = 0.05
    if p_val < 0.05:
        final_diagnosis = "Reject H0: Baseline center has drifted. First moment is heterogenous."
    else:
        final_diagnosis = "Fail to Reject H0: Baseline is stable. First moment is homogenous."

    return {
        "Wilks_Lambda": round(lambda_wilks, 5),
        "Bartlett_Chi2_Stat": round(chi2_stat, 3),
        "Degrees_of_Freedom": dof,
        "P_Value": p_val,
        "Diagnostic_Conclusion": final_diagnosis
    }

# Execute script
results = verify_first_moment_homogeneity(df_strain, g_chunks=5)
for metric, value in results.items():
    print(f"{metric}: {value}")
```
