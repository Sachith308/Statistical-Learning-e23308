# Q. Analytical Derivation

Consider the filter model
\begin{align*}
x^{-}_k &= A_{k-1}\,x^{+}_{k-1} + G_{k-1}\,w_{k-1}, \\
y^{-}_k &= H_k\,x^{-}_k + z_k,
\end{align*}
where
$$x^{+}_{k-1} \sim \mathscr{N}(m_{k-1}, P_{k-1}),$$

Answer the followings:

1. Show that
$$x_k^- \sim \mathscr{N}(m_k^-, P_k^-),$$
where
$$m_k^- = A_{k-1}m_{k-1},$$
and
$$
P_k^- = A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T.
$$

2. Show that
$$y_k^- \sim \mathscr{N}(H_k m_k^-, H_k P_k^- H_k^T + \Sigma_m).$$

3. Show that
$$\begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix} \sim \mathscr{N}\left(
\begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix},
\begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}
\right).$$

4. Show that
$$x^{+}_k\triangleq (x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}) \sim \mathscr{N}(m_k, P_k),$$
where the updated mean $m_k$ and updated covariance $P_k$ are given by:
\begin{align*}
K_k &\triangleq P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}, \\
m_k &= m_k^- + K_k (y^{\mathrm{obs}}_{k} - H_k m_k^-), \\
P_k &= (I - K_k H_k) P_k^-.
\end{align*}

5. Find
$\mathbb{E}[x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}]$  and
$\text{Var}(x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k})$.

## Answers
### 1)

Given
$$
[
x_k^- = A_{k-1}x_{k-1}^+ + G_{k-1}w_{k-1},
]
$$
where
$$
[
x_{k-1}^+ \sim \mathcal N(m_{k-1},P_{k-1}),
]
$$
and
$$
[
w_{k-1}\sim\mathcal N(0,\Sigma_p).
]
$$
Since a linear transformation of a Gaussian random vector is also Gaussian,
$$
[
x_k^- \sim \mathcal N(m_k^-,P_k^-).
]
$$

The mean is
$$[
\begin{aligned}
m_k^-
&=\mathbb E[x_k^-]\\
&=\mathbb E[A_{k-1}x_{k-1}^+ + G_{k-1}w_{k-1}] \\
&=A_{k-1}\mathbb E[x_{k-1}^+]
+G_{k-1}\mathbb E[w_{k-1}] \\
&=A_{k-1}m_{k-1},
\end{aligned}]
$$
because
$$
\mathbb E[w_{k-1}]=0.
$$
The covariance is
$$
[
\begin{aligned}
P_k^-
&=\operatorname{Var}(x_k^-)\\
&=\operatorname{Var}(A_{k-1}x_{k-1}^+
+G_{k-1}w_{k-1})\\
&=A_{k-1}P_{k-1}A_{k-1}^T
+
G_{k-1}\Sigma_pG_{k-1}^T,
\end{aligned}
]
$$
since the state and process noise are independent.

Hence,
$$
[
\boxed{
x_k^-
\sim
\mathcal N
(m_k^-,P_k^-)
}
]
$$

### 2)

The measurement model is
$$
[
y_k^-=H_kx_k^-+z_k,
]
$$
where
$$
[
z_k\sim\mathcal N(0,\Sigma_m).
]
$$
$$
Since (y_k^-) is a linear transformation of Gaussian variables,
$$
$$
[
y_k^-
\sim
\mathcal N
(H_km_k^-,
H_kP_k^-H_k^T+\Sigma_m).
]
$$

The mean is
$$
[
\begin{aligned}
\mathbb E[y_k^-]
&=
\mathbb E[H_kx_k^-+z_k] \\
&=
H_k\mathbb E[x_k^-]
+\mathbb E[z_k] \\
&=
H_km_k^-.
\end{aligned}
]
$$

The covariance is
$$
[
\begin{aligned}
\operatorname{Var}(y_k^-)
&=
\operatorname{Var}(H_kx_k^-+z_k)\\
&=
H_kP_k^-H_k^T+\Sigma_m.
\end{aligned}
]
$$
Therefore,
$$
[
\boxed{
y_k^-
\sim
\mathcal N
(H_km_k^-,
H_kP_k^-H_k^T+\Sigma_m)
}
]
$$

### 3)

The vector
$$

\begin{bmatrix}
x_k^-\\
y_k^-
\end{bmatrix}

$$
is jointly Gaussian because both variables are linear functions of Gaussian random vectors.

The joint mean is
$$
[
\begin{aligned}
\mathbb E
\left[
\begin{bmatrix}
x_k^-\\
y_k^-
\end{bmatrix}
\right]
=
\begin{bmatrix}
m_k^-\\
H_km_k^-
\end{bmatrix}.
\end{aligned}
]
$$
The covariance matrix is
$$
[
\begin{aligned}
\operatorname{Cov}
\left(
\begin{bmatrix}
x_k^-\\
y_k^-
\end{bmatrix}
\right)
=
\begin{bmatrix}
P_k^- &
P_k^-H_k^T\\
H_kP_k^- &
H_kP_k^-H_k^T+\Sigma_m
\end{bmatrix}.
\end{aligned}
]
$$

Hence,
$$
[
\boxed{
\begin{bmatrix}
x_k^-\\
y_k^-
\end{bmatrix}
\sim
\mathcal N
\left(
\begin{bmatrix}
m_k^-\\
H_km_k^-
\end{bmatrix},
\begin{bmatrix}
P_k^-&
P_k^-H_k^T\\
H_kP_k^-&
H_kP_k^-H_k^T+\Sigma_m
\end{bmatrix}
\right)
}
]
$$

### 4)

Using the conditional distribution of jointly Gaussian random variables, the posterior state is
$$
[
x_k^+
=
(x_k^-|y_k^-=y_k^{obs})
\sim
\mathcal N(m_k,P_k).
]
$$
The Kalman gain is
$$
[
\boxed{
K_k
=
P_k^-H_k^T
(H_kP_k^-H_k^T+\Sigma_m)^{-1}
}
]
$$

The updated mean is
$$
[
\boxed{
m_k
=
m_k^-
+
K_k
(y_k^{obs}-H_km_k^-)
}
]
$$

The updated covariance is
$$
[
\boxed{
P_k
=
(I-K_kH_k)P_k^-.
}
]$$

Therefore,
$$
[
\boxed{
x_k^+
\sim
\mathcal N(m_k,P_k)
}
]
$$

### 5)

The conditional expectation of the predicted state is equal to the posterior mean.

Therefore,
$$
[
\boxed{
\mathbb E
[x_k^-|y_k^-=y_k^{obs}]
=
m_k
}
]
$$

Substituting the expression for \(m_k\),
$$
[
\boxed{
\mathbb E
[x_k^-|y_k^-=y_k^{obs}]
=
m_k^-
+
K_k
(y_k^{obs}-H_km_k^-)
}
]
$$
Similarly,
$$
[
\boxed{
\operatorname{Var}
(x_k^-|y_k^-=y_k^{obs})
=
P_k
}
]
$$

or
$$
[
\boxed{
\operatorname{Var}
(x_k^-|y_k^-=y_k^{obs})
=
(I-K_kH_k)P_k^-.
}
]
$$

# Q. 1-D Example

Consider the scalar linear-Gaussian filter model
\begin{aligned}
x^-_k &= a\,x^+_{k-1} + w_{k-1},\qquad w_{k-1}\sim\mathscr N(0,q),\\
y^-_k &= h\,x^-_k + z_k,\qquad\;\;\;\; z_k\sim\mathscr N(0,r),
\end{aligned}
with prior $x_0\sim \mathscr{N}(m_0,P_0)$. Define $Y_{k}=\{y^{\mathrm{obs}}_1,\dots,y^{\mathrm{obs}}_{k}\}$.

Answer the following:

1. Show that
\begin{aligned}
m_k^- &= a\,m_{k-1},\\
P_k^- &= a^2 P_{k-1} + q.
\end{aligned}

2. Show that
\begin{aligned}
m_k &= m_k^- + K_k\,v_k
= m_k^- + \frac{P_k^- h}{S_k}\bigl(y^{\mathrm{obs}}_k - h\,m_k^-\bigr),\\
P_k &= (1 - K_k h)\,P_k^-
= \Bigl(1 - \frac{P_k^- h^2}{S_k}\Bigr) P_k^- .
\end{aligned}

3. Show that the predictive measurement distribution (before seeing $y_k$):
\begin{align*}
p(y^-_k | Y_{k-1})=\mathscr N\bigl(h\,m_k^-,\; h^2 P_k^- + r\bigr).
\end{align*}

4. Show that the posterior-predictive measurement distribution (after filtering on $y_k$):
\begin{align*}
p(y^-_k\mid Y_k)=\mathscr N\bigl(h\,m_k,\; h^2 P_k + r\bigr).
\end{align*}

5. For suitable numerical values for $a,q,h,r$ animate the above to prior distribution and the posterior distribution.

## Answers

### 1)

The state equation is
$$
[
x_k^- = ax_{k-1}^+ + w_{k-1},
]
$$

where
$$
[
w_{k-1}\sim\mathcal N(0,q).
]
$$

Since the state and process noise are Gaussian,
$$
[
x_k^- \sim \mathcal N(m_k^-,P_k^-)
]
$$

The predicted mean is
$$
[
\begin{aligned}
m_k^-
&=\mathbb E[x_k^-] \\
&=\mathbb E[ax_{k-1}^+ + w_{k-1}]\\
&=a\mathbb E[x_{k-1}^+] + \mathbb E[w_{k-1}]\\
&=am_{k-1}.
\end{aligned}
]
$$

The predicted covariance is
$$
[
\begin{aligned}
P_k^-
&=\operatorname{Var}(x_k^-)\\
&=\operatorname{Var}(ax_{k-1}^+ + w_{k-1})\\
&=a^2P_{k-1}+q.
\end{aligned}
]
$$
Hence,
$$
[
\boxed{
m_k^-=am_{k-1},
\qquad
P_k^-=a^2P_{k-1}+q.
}
]
$$

### 2)

The innovation is
$$
[
v_k
=
y_k^{obs}-hm_k^-.
]
$$

The innovation covariance is
$$
[
S_k=h^2P_k^-+r.
]
$$

The Kalman gain is
$$
[
K_k=\frac{P_k^-h}{S_k}.
]
$$

Therefore, the updated mean is
$$
[
\boxed{
m_k
=
m_k^-+K_kv_k
}
]
$$

or
$$
[
m_k
=
m_k^-
+
\frac{P_k^-h}{S_k}
(y_k^{obs}-hm_k^-).
]
$$

The updated covariance is
$$
[
\boxed{
P_k=(1-K_kh)P_k^-.
}
]
$$

### 3)

The measurement equation is
$$
[
y_k
=
hx_k^-+z_k,
]
$$

where
$$
[
z_k
\sim
\mathcal N(0,r).
]
$$
The predictive mean is
$$
[
\mathbb E[y_k]
=
hm_k^-.
]
$$
The predictive variance is
$$
[
\operatorname{Var}(y_k)
=
h^2P_k^-+r.
]
$$
Therefore,
$$
[
\boxed{
p(y_k|Y_{k-1})
=
\mathcal N
(hm_k^-,
h^2P_k^-+r).
}
]
$$

### 4)

After incorporating the measurement,
$$
[
x_k^+
\sim
\mathcal N(m_k,P_k).
]
$$
The measurement equation remains
$$
[
y_k
=
hx_k^+
+
z_k.
]
$$
The posterior predictive mean is
$$
[
hm_k.
]
$$
The posterior predictive variance is
$$
[
h^2P_k+r.
]
$$
Hence,
$$
[
\boxed{
p(y_k|Y_k)
=
\mathcal N
(hm_k,
h^2P_k+r).
}
]
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
# System Parameters
a = 1.0       # State transition
h = 1.0       # Measurement matrix
q = 0.2       # Process noise variance
r = 1.0       # Measurement noise variance

# Number of time steps
N = 30

# Initial estimate
m = 0.0
P = 1.0
```

```python
np.random.seed(42)

x_true = np.zeros(N)
y = np.zeros(N)

x_true[0] = 0

for k in range(1, N):
    x_true[k] = a*x_true[k-1] + np.random.normal(0, np.sqrt(q))

for k in range(N):
    y[k] = h*x_true[k] + np.random.normal(0, np.sqrt(r))
```

```python
m_prior = []
P_prior = []

m_post = []
P_post = []

for k in range(N):

    # Prediction
    m_minus = a*m
    P_minus = a*a*P + q

    # Kalman Gain
    S = h*h*P_minus + r
    K = P_minus*h/S

    # Update
    innovation = y[k] - h*m_minus
    m = m_minus + K*innovation
    P = (1-K*h)*P_minus

    m_prior.append(m_minus)
    P_prior.append(P_minus)

    m_post.append(m)
    P_post.append(P)
```

```python
plt.figure(figsize=(10,6))

plt.plot(x_true,'k',label='True State')
plt.plot(y,'rx',label='Measurements')
plt.plot(m_prior,'b--',label='Prior Mean')
plt.plot(m_post,'g',label='Posterior Mean')

plt.xlabel("Time Step")
plt.ylabel("State")
plt.title("1-D Kalman Filter")

plt.legend()
plt.grid(True)

plt.show()
```

```python
k = N-1

x = np.linspace(-5,5,400)

prior_pdf = norm.pdf(
    x,
    m_prior[k],
    np.sqrt(P_prior[k])
)

posterior_pdf = norm.pdf(
    x,
    m_post[k],
    np.sqrt(P_post[k])
)

plt.figure(figsize=(8,5))

plt.plot(x,prior_pdf,label='Prior')
plt.plot(x,posterior_pdf,label='Posterior')

plt.title("Prior and Posterior Distributions")

plt.xlabel("State")
plt.ylabel("Probability Density")

plt.legend()

plt.grid(True)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))

x = np.linspace(-5,5,400)

line1, = ax.plot([],[],label='Prior')
line2, = ax.plot([],[],label='Posterior')

ax.set_xlim(-5,5)
ax.set_ylim(0,1.5)

ax.set_xlabel("State")
ax.set_ylabel("Probability Density")

ax.legend()

def animate(i):

    prior = norm.pdf(
        x,
        m_prior[i],
        np.sqrt(P_prior[i])
    )

    posterior = norm.pdf(
        x,
        m_post[i],
        np.sqrt(P_post[i])
    )

    line1.set_data(x,prior)
    line2.set_data(x,posterior)

    ax.set_title(f"Time Step {i}")

    return line1,line2

ani = FuncAnimation(
    fig,
    animate,
    frames=N,
    interval=500,
    blit=True
)

HTML(ani.to_jshtml())
```

```python
print("Final Prior Mean      :", m_prior[-1])
print("Final Prior Variance  :", P_prior[-1])

print()

print("Final Posterior Mean     :", m_post[-1])
print("Final Posterior Variance :", P_post[-1])
```

## Part B

Develop a python based method to filter a sequence of noisy GPS obtained position meaurements.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 1.0

A = np.array([
    [1,0,dt,0],
    [0,1,0,dt],
    [0,0,1,0],
    [0,0,0,1]
])

G = np.array([
    [0.5*dt**2,0],
    [0,0.5*dt**2],
    [dt,0],
    [0,dt]
])

H = np.array([
    [1,0,0,0],
    [0,1,0,0]
])
```

```python
Sigma_p = np.diag([0.2,0.2])

Sigma_m = np.diag([2.0,2.0])

Q = G @ Sigma_p @ G.T

R = Sigma_m
```

```python
N = 100

x_true = np.zeros((4,N))

x_true[:,0] = [0,0,1,0.5]

for k in range(1,N):

    noise = np.random.multivariate_normal(
        np.zeros(2),
        Sigma_p
    )

    x_true[:,k] = (
        A @ x_true[:,k-1]
        + G @ noise
    )
```

```python
gps = np.zeros((2,N))

for k in range(N):

    noise = np.random.multivariate_normal(
        np.zeros(2),
        Sigma_m
    )

    gps[:,k] = (
        H @ x_true[:,k]
        + noise
    )
```

```python
x = np.zeros((4,1))

P = np.eye(4)

filtered = np.zeros((4,N))
```

```python
for k in range(N):

    # Prediction

    x_pred = A @ x

    P_pred = A @ P @ A.T + Q

    # Kalman Gain

    S = H @ P_pred @ H.T + R

    K = P_pred @ H.T @ np.linalg.inv(S)

    # Update

    z = gps[:,k].reshape(2,1)

    x = x_pred + K @ (z - H @ x_pred)

    P = (np.eye(4)-K@H) @ P_pred

    filtered[:,k] = x.flatten()
```

```python
plt.figure(figsize=(8,8))

plt.plot(
    x_true[0],
    x_true[1],
    label="True Path",
    linewidth=2
)

plt.scatter(
    gps[0],
    gps[1],
    s=15,
    label="GPS"
)

plt.plot(
    filtered[0],
    filtered[1],
    linewidth=2,
    label="Kalman Filter"
)

plt.xlabel("X Position")

plt.ylabel("Y Position")

plt.title("2D Position Estimation")

plt.legend()

plt.grid()

plt.show()
```

```python
gps_error = np.mean(
    np.sqrt(
        (gps[0]-x_true[0])**2 +
        (gps[1]-x_true[1])**2
    )
)

kf_error = np.mean(
    np.sqrt(
        (filtered[0]-x_true[0])**2 +
        (filtered[1]-x_true[1])**2
    )
)

print("Average GPS Error :", gps_error)

print("Average Kalman Filter Error :", kf_error)
```
