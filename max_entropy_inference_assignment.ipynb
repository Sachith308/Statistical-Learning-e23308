# Q1. Statistical Mechanics of the Quantum Oscillator

Consider a single quantum harmonic oscillator of angular frequency $\omega$ in thermal equilibrium with a heat bath at temperature $T$. The quantum mechanical energy eigenvalues are given by:

$$E_j = \hbar \omega \left(j + \frac{1}{2}\right), \quad \text{for } j = 0, 1, 2, \dots$$

where $\hbar$ is the reduced Planck constant and $k_B$ is the Boltzmann constant.

**Answer the Following:**

1. **The Partition Function:**
Show that the canonical partition function $Z(T) = \sum_{j=0}^\infty e^{-E_j / (k_B T)}$ forms a convergent geometric series that simplifies to:

$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}}$$


2. **Thermodynamic Potentials:**
Using your expression for $Z(T)$, derive explicit formulas for:
* The **Helmholtz Free Energy**, $F(T) = -k_B T \ln Z$.
* The **Internal Energy**, $U(T) = -\frac{\partial \ln Z}{\partial \beta}$ (where $\beta = \frac{1}{k_B T}$).


3. **Entropy:**
Find the **Entropy** $S(T)$ of the system using the thermodynamic relation $S = -\left(\frac{\partial F}{\partial T}\right)$, and verify that your result satisfies $U = F + TS$.
4. **Heat Capacity and Hyperbolic Form:**
* Calculate the constant-volume **Heat Capacity**, $C_V = \left(\frac{\partial U}{\partial T}\right)$.
* Use the hyperbolic identity $\sinh(x) = \frac{e^x - e^{-x}}{2}$ to prove that your expression for $C_V$ can be compactly rewritten as:

$$C_V = k_B \left(\frac{\hbar \omega}{2k_BT}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_BT}\right)}$$




5. **Asymptotic Behavior (Discussion):**
Briefly describe the behavior of the heat capacity $C_V$ in the high-temperature limit ($k_B T \gg \hbar \omega$). Does it match the classical equipartition theorem prediction?

# Q1. Answer

The energy eigenvalues of a single quantum harmonic oscillator are

$$
E_j=\hbar\omega\left(j+\frac12\right),
\qquad j=0,1,2,\ldots
$$

where

- $\hbar$ is the reduced Planck constant,
- $\omega$ is the angular frequency,
- $k_B$ is the Boltzmann constant,
- $T$ is the absolute temperature.

Define

$$
\beta=\frac{1}{k_BT}.
$$

---

## 1. Partition Function

The canonical partition function is

$$
Z=\sum_{j=0}^{\infty}e^{-\beta E_j}.
$$

Substituting the energy eigenvalues,

$$
\begin{aligned}
Z
&=\sum_{j=0}^{\infty}
e^{-\beta\hbar\omega\left(j+\frac12\right)} \\
&=
e^{-\beta\hbar\omega/2}
\sum_{j=0}^{\infty}
\left(e^{-\beta\hbar\omega}\right)^j.
\end{aligned}
$$

Since

$$
0<e^{-\beta\hbar\omega}<1,
$$

the infinite geometric series converges,

$$
\sum_{j=0}^{\infty}r^j=\frac1{1-r}.
$$

Therefore,

$$
\boxed{
Z=
\frac{
e^{-\beta\hbar\omega/2}
}{
1-e^{-\beta\hbar\omega}
}
}
$$

---

# 2. Thermodynamic Potentials

## (a) Helmholtz Free Energy

The Helmholtz free energy is

$$
F=-k_BT\ln Z.
$$

Using the expression for $Z$,

$$
\ln Z
=
-\frac{\beta\hbar\omega}{2}
-\ln\left(1-e^{-\beta\hbar\omega}\right).
$$

Hence,

$$
\begin{aligned}
F
&=
-k_BT
\left[
-\frac{\beta\hbar\omega}{2}
-\ln\left(1-e^{-\beta\hbar\omega}\right)
\right] \\
&=
\frac{\hbar\omega}{2}
+
k_BT
\ln\left(1-e^{-\hbar\omega/(k_BT)}\right).
\end{aligned}
$$

Therefore,

$$
\boxed{
F=
\frac{\hbar\omega}{2}
+
k_BT
\ln\left(1-e^{-\hbar\omega/(k_BT)}\right)
}
$$

---

## (b) Internal Energy

The internal energy is

$$
U
=
-\frac{\partial\ln Z}{\partial\beta}.
$$

Differentiate

$$
\ln Z
=
-\frac{\beta\hbar\omega}{2}
-
\ln\left(1-e^{-\beta\hbar\omega}\right).
$$

Then,

$$
\frac{\partial\ln Z}{\partial\beta}
=
-\frac{\hbar\omega}{2}
-
\frac{\hbar\omega e^{-\beta\hbar\omega}}
{1-e^{-\beta\hbar\omega}}.
$$

Thus,

$$
U
=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}
{e^{\beta\hbar\omega}-1}.
$$

Replacing $\beta=1/(k_BT)$,

$$
\boxed{
U=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}
{e^{\hbar\omega/(k_BT)}-1}
}
$$

---

# 3. Entropy

The entropy is

$$
S
=
-\left(\frac{\partial F}{\partial T}\right).
$$

A convenient equivalent expression is

$$
S=\frac{U-F}{T}.
$$

Substituting the expressions for $U$ and $F$,

$$
S
=
\frac1T
\left[
\frac{\hbar\omega}
{e^{\hbar\omega/(k_BT)}-1}
-
k_BT
\ln\left(
1-e^{-\hbar\omega/(k_BT)}
\right)
\right].
$$

Therefore,

$$
\boxed{
S
=
\frac{\hbar\omega}
{T\left(e^{\hbar\omega/(k_BT)}-1\right)}
-
k_B
\ln\left(
1-e^{-\hbar\omega/(k_BT)}
\right)
}
$$

### Verification

Using

$$
F+TS,
$$

we obtain

$$
\begin{aligned}
F+TS
&=
\frac{\hbar\omega}{2}
+
k_BT\ln(1-e^{-x}) \\
&\quad+
\frac{\hbar\omega}
{e^x-1}
-
k_BT\ln(1-e^{-x}) \\
&=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}
{e^x-1},
\end{aligned}
$$

where

$$
x=\frac{\hbar\omega}{k_BT}.
$$

Hence,

$$
\boxed{
U=F+TS
}
$$

which verifies the thermodynamic identity.

---

# 4. Heat Capacity

The constant-volume heat capacity is

$$
C_V
=
\left(
\frac{\partial U}{\partial T}
\right)_V.
$$

Let

$$
x=\frac{\hbar\omega}{k_BT}.
$$

Then

$$
U
=
\frac{\hbar\omega}{2}
+
\frac{\hbar\omega}{e^x-1}.
$$

Since

$$
\frac{dx}{dT}
=
-\frac{x}{T},
$$

differentiating gives

$$
\boxed{
C_V
=
k_B
\left(
\frac{\hbar\omega}{k_BT}
\right)^2
\frac{e^x}
{(e^x-1)^2}
}
$$

---

## Hyperbolic Form

Using

$$
\sinh(y)
=
\frac{e^y-e^{-y}}2,
$$

let

$$
y=\frac{x}{2}.
$$

Then,

$$
\frac{e^x}{(e^x-1)^2}
=
\frac1{4\sinh^2(x/2)}.
$$

Hence,

$$
\boxed{
C_V
=
k_B
\left(
\frac{\hbar\omega}{2k_BT}
\right)^2
\frac1
{\sinh^2\left(
\frac{\hbar\omega}{2k_BT}
\right)}
}
$$

---

# 5. High-Temperature Limit

When

$$
k_BT\gg\hbar\omega,
$$

we have

$$
x=\frac{\hbar\omega}{k_BT}\ll1.
$$

Using the approximation

$$
\sinh(x/2)\approx\frac{x}{2},
$$

the heat capacity becomes

$$
C_V
\approx
k_B
\left(\frac{x}{2}\right)^2
\frac1{(x/2)^2}
=
k_B.
$$

Therefore,

$$
\boxed{
\lim_{T\to\infty}C_V=k_B
}
$$

This agrees with the **classical equipartition theorem**, which predicts that a one-dimensional harmonic oscillator contributes exactly one Boltzmann constant, $k_B$, to the heat capacity.

# Q2. The Maximum Entropy Distribution under Mean and Variance Constraints

Consider the random variable $X \in (-\infty, \infty)$. Let:

1. **The Mean:** $\mathbb{E}[X] = \mu$
2. **The Variance:** $\mathbb{E}[(X-\mu)^2] = \sigma^2$

Show that the law of $X$ that maximizes entropy is given by:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

# Q2. Answers

Consider a continuous random variable

$$
X\in(-\infty,\infty),
$$

with the following constraints:

1. **Normalization**

$$
\int_{-\infty}^{\infty}p(x)\,dx=1.
$$

2. **Mean**

$$
E[X]=\int_{-\infty}^{\infty}xp(x)\,dx=\mu.
$$

3. **Variance**

$$
E[(X-\mu)^2]
=
\int_{-\infty}^{\infty}(x-\mu)^2p(x)\,dx
=
\sigma^2.
$$

Our objective is to determine the probability density function that maximizes the Shannon differential entropy.

# Define the Entropy

The differential entropy of a continuous distribution is

$$
H[p]
=
-\int_{-\infty}^{\infty}
p(x)\ln p(x)\,dx.
$$

We seek the density \(p(x)\) that maximizes \(H[p]\) subject to the given constraints.



# Construct the Lagrangian

Introduce the Lagrange multipliers

- $\lambda_0$ for normalization,
- $\lambda_1$ for the mean,
- $\lambda_2$ for the variance.

The Lagrangian functional becomes

$$
\begin{aligned}
\mathcal{L}
=&
-
\int p(x)\ln p(x)\,dx
\\
&
-\lambda_0
\left(
\int p(x)\,dx-1
\right)
\\
&
-\lambda_1
\left(
\int xp(x)\,dx-\mu
\right)
\\
&
-\lambda_2
\left(
\int (x-\mu)^2p(x)\,dx-\sigma^2
\right).
\end{aligned}
$$


# Apply the Euler Lagrange Condition

To maximize the entropy,

$$
\frac{\delta\mathcal{L}}{\delta p(x)}=0.
$$

Differentiating with respect to \(p(x)\),

$$
-\ln p(x)-1
-\lambda_0
-\lambda_1x
-\lambda_2(x-\mu)^2
=0.
$$

Hence,

$$
\ln p(x)
=
-1-\lambda_0-\lambda_1x-\lambda_2(x-\mu)^2.
$$

Exponentiating both sides,

$$
p(x)
=
\exp
\left(
-1-\lambda_0
-\lambda_1x
-\lambda_2(x-\mu)^2
\right).
$$

Define

$$
A=e^{-1-\lambda_0},
$$

so that

$$
\boxed{
p(x)
=
A
\exp
\left(
-\lambda_1x
-\lambda_2(x-\mu)^2
\right).
}
$$

# Determine the Constants

Because the prescribed mean is exactly \(\mu\), symmetry requires

$$
\boxed{\lambda_1=0.}
$$

Thus,

$$
p(x)
=
A
e^{-\lambda_2(x-\mu)^2}.
$$

The normalization condition gives

$$
1
=
A
\int_{-\infty}^{\infty}
e^{-\lambda_2(x-\mu)^2}
dx.
$$

Using the Gaussian integral

$$
\int_{-\infty}^{\infty}
e^{-ax^2}
dx
=
\sqrt{\frac{\pi}{a}},
\qquad a>0,
$$

we obtain

$$
A
=
\sqrt{\frac{\lambda_2}{\pi}}.
$$

Next, use the variance constraint,

$$
\sigma^2
=
\int
(x-\mu)^2
p(x)\,dx.
$$

Evaluating the Gaussian integral,

$$
\boxed{
\lambda_2
=
\frac1{2\sigma^2}.
}
$$

Therefore,

$$
A
=
\frac1{\sqrt{2\pi\sigma^2}}.
$$

# Maximum Entropy Distribution

Substituting the constants,

$$
\boxed{
p(x)
=
\frac1{\sqrt{2\pi\sigma^2}}
\exp
\left(
-\frac{(x-\mu)^2}
{2\sigma^2}
\right).
}
$$

Hence,

$$
\boxed{
X
\sim
\mathcal N(\mu,\sigma^2).
}
$$

# Result

The probability density function with the **maximum entropy** among all continuous distributions having the specified mean and variance is the Gaussian distribution,

$$
\boxed{
p(x)
=
\frac1{\sqrt{2\pi\sigma^2}}
\exp
\left(
-\frac{(x-\mu)^2}
{2\sigma^2}
\right).
}
$$

Thus, the **Normal distribution is the unique maximum entropy distribution** under fixed mean and variance constraints.

# Q3. Answers

Consider the binary classification problem where

$$
Y\in\{0,1\},
$$

with

- \(Y=0\): Finance
- \(Y=1\): Technical

The feature vector is

$$
x=(x_1,x_2,x_3)^T,
$$

where

The observed feature constraints are

$$

$$

$$

$$
E[x_3I(Y=1)] = 0.16.
$$

The marginal probabilities of the feature vectors are fixed by historical observations.

# Maximum Entropy Principle

The conditional entropy is

$$
H(Y|X)
=
-
\sum_x
P(x)
\sum_y
P(y|x)
\ln P(y|x).
$$

The objective is

$$
\boxed{
\max_{P(y|x)} H(Y|X)
}
$$

subject to the observed feature expectations.

The conditional probabilities must satisfy

### (i) Normalization

For every feature vector,

$$
\sum_y P(y|x)=1.
$$

### (ii) Feature Constraints

For each feature,

$$
\sum_x
P(x)
P(Y=1|x)
x_1
=
0.05,
$$

$$
\sum_x
P(x)
P(Y=1|x)
x_2
=
0.15,
$$

$$
\sum_x
P(x)
P(Y=1|x)
x_3
=
0.16.
$$

# Form the Lagrangian

Introduce the Lagrange multipliers

$$
\lambda_1,\lambda_2,\lambda_3
$$

for the feature constraints and

$$
\alpha(x)
$$

for normalization.

The Lagrangian is

$$
\begin{aligned}
\mathcal L
=&
-
\sum_x
P(x)
\sum_y
P(y|x)
\ln P(y|x)
\\
&
+
\sum_{i=1}^{3}
\lambda_i
\left(
E[f_i]-c_i
\right)
\\
&
+
\sum_x
\alpha(x)
\left(
\sum_yP(y|x)-1
\right),
\end{aligned}
$$

where

$$
f_i(x,y)=x_iI(y=1).
$$

# Differentiate

Taking the derivative with respect to

$$
P(y|x)
$$

gives

$$
-\ln P(y|x)-1
+
\sum_i
\lambda_i
f_i(x,y)
+
\alpha(x)
=
0.
$$

Therefore,

$$
P(y|x)
=
\exp
\left(
\alpha(x)-1+
\sum_i
\lambda_i
f_i(x,y)
\right).
$$

# Normalize

Using

$$
\sum_yP(y|x)=1,
$$

the normalization constant becomes

$$
Z(x)
=
\sum_y
\exp
\left(
\sum_i
\lambda_i
f_i(x,y)
\right).
$$

Hence,

$$
\boxed{
P(y|x)
=
\frac{
\exp\left(\sum_i\lambda_if_i(x,y)\right)

Z(x)
}
}
$$

This is the Maximum Entropy model.

$$

we have

}
}
$$

Similarly,

The Maximum Entropy solution is the logistic regression model

P(Y=1|x)

# Q4. Asset Pricing and Portfolio Optimization.



In standard finance, models often assume asset returns follow a perfect Normal (Gaussian) distribution. However, real-world markets experience extreme events, leading to "fat tails" and skewed returns. MaxEnt allows an analyst to infer the **least-biased probability distribution** of an asset's future price based *only* on what the market currently tells us, without forcing it into a rigid Gaussian shape.

**Background:**

You are a quantitative portfolio manager evaluating a highly volatile tech stock, **Asset X**, currently trading at $\$100$. You want to determine the risk-neutral probability distribution of its price, $S_T$, in exactly one month. This distribution will allow you to correctly price exotic derivatives and optimize your portfolio's risk exposure.

**Available Market Data (The Constraints):**

You do not know the true distribution, but you can observe the following macroeconomic aggregates and market-clearing prices from the options exchange:

1. **The Risk-Free Growth:** Based on the risk-free interest rate, the expected future price of the asset must match the forward price:

$$\mathbb{E}[S_T] = \$102$$


2. **The Market Volatility:** Based on standard historical variance, the expected squared price is constrained by:

$$\mathbb{E}[S_T^2] = 10,600$$


3. **Asymmetric Market Sentiment (The "Skew"):** Due to heavy trading on out-of-the-money put options, the market is pricing in a strong downside risk. The expected value of a specific logarithmic penalty (which captures downward crash risk) is constrained to:

$$\mathbb{E}[\ln(S_T)] = 4.60$$



**Objective:**

Formulate and solve the Maximum Entropy problem to find the continuous probability density function $p(S_T)$ for $S_T > 0$.



# Q4. Answers

Let \(S_T > 0\) denote the asset price after one month. We wish to determine the probability density function \(p(S_T)\) that maximizes the entropy while satisfying the observed market constraints.

The available information is

$$
E[S_T]=102,
$$

$$
E[S_T^2]=10600,
$$

and

$$
E[\ln S_T]=4.60.
$$

# Define the Entropy

The differential entropy of a continuous distribution is

$$
H[p]
=
-
\int_{0}^{\infty}
p(S_T)
\ln p(S_T)
\,dS_T.
$$

Our objective is

$$
\boxed{
\max_{p(S_T)}
H[p].
}
$$

# Constraints

The probability density must satisfy the following constraints.

### (a) Normalization

$$
\int_0^{\infty}
p(S_T)\,dS_T
=
1.
$$

### (b) Expected Price

$$
\int_0^{\infty}
S_T\,p(S_T)\,dS_T
=
102.
$$

### (c) Expected Squared Price

$$
\int_0^{\infty}
S_T^2\,p(S_T)\,dS_T
=
10600.
$$

### (d) Expected Logarithmic Price

$$
\int_0^{\infty}
\ln(S_T)\,
p(S_T)\,dS_T
=
4.60.
$$

# Construct the Lagrangian

Introduce the Lagrange multipliers

$$
\lambda_0,\;
\lambda_1,\;
\lambda_2,\;
\lambda_3.
$$

The Lagrangian functional becomes

$$
\begin{aligned}
\mathcal L
=&
-
\int
p(S_T)\ln p(S_T)\,dS_T
\\
&
-\lambda_0
\left(
\int p(S_T)\,dS_T-1
\right)
\\
&
-\lambda_1
\left(
\int S_Tp(S_T)\,dS_T-102
\right)
\\
&
-\lambda_2
\left(
\int S_T^2p(S_T)\,dS_T-10600
\right)
\\
&
-\lambda_3
\left(
\int
\ln(S_T)
p(S_T)\,dS_T
-
4.60
\right).
\end{aligned}
$$

# Differentiate

To maximize entropy,

$$
\frac{\delta\mathcal L}{\delta p(S_T)}
=
0.
$$

Differentiating gives

$$
-\ln p(S_T)-1
-\lambda_0
-\lambda_1S_T
-\lambda_2S_T^2
-\lambda_3\ln(S_T)
=
0.
$$

Therefore,

$$
\ln p(S_T)
=
-1
-\lambda_0
-\lambda_1S_T
-\lambda_2S_T^2
-\lambda_3\ln(S_T).
$$

Exponentiating both sides,

$$
p(S_T)
=
e^{-1-\lambda_0}
S_T^{-\lambda_3}
e^{-\lambda_1S_T}
e^{-\lambda_2S_T^2}.
$$

Define

$$
A=e^{-1-\lambda_0},
$$

to obtain

$$
\boxed{
p(S_T)
=
A
S_T^{-\lambda_3}
\exp
\left(
-\lambda_1S_T
-\lambda_2S_T^2
\right).
}
$$

# Determine the Normalization Constant

The constant \(A\) is chosen so that

$$
\int_0^{\infty}
p(S_T)\,dS_T
=
1.
$$

Hence,

$$
A
=
\left[
\int_0^{\infty}
S_T^{-\lambda_3}
e^{-\lambda_1S_T-\lambda_2S_T^2}
\,dS_T
\right]^{-1}.
$$

The remaining Lagrange multipliers

$$
\lambda_1,\lambda_2,\lambda_3
$$

are determined numerically so that

$$
E[S_T]=102,
$$

$$
E[S_T^2]=10600,
$$

and

$$
E[\ln S_T]=4.60.
$$

# Interpretation

Each constraint contributes one term to the exponential-family density.

- The **first-moment constraint** introduces the linear term

$$
e^{-\lambda_1S_T}.
$$

- The **second-moment constraint** introduces the quadratic term

$$
e^{-\lambda_2S_T^2},
$$

which controls the dispersion of the distribution.

- The **logarithmic constraint** introduces the power-law factor

$$
S_T^{-\lambda_3},
$$

allowing the model to represent asymmetric and heavy-tailed market behavior.

Unlike the Gaussian distribution, this Maximum Entropy solution can better capture skewness and crash risk observed in financial markets.

# Result

The Maximum Entropy probability density satisfying all the market constraints is

$$
\boxed{
p(S_T)
=
A
S_T^{-\lambda_3}
\exp
\left(
-\lambda_1S_T
-\lambda_2S_T^2
\right),
\qquad S_T>0,
}
$$

where

$$
A
=
\left[
\int_0^{\infty}
S_T^{-\lambda_3}
e^{-\lambda_1S_T-\lambda_2S_T^2}
\,dS_T
\right]^{-1},
$$

and the parameters

$$
\lambda_1,\lambda_2,\lambda_3
$$

are obtained by enforcing the three market constraints:

$$
E[S_T]=102,
$$

$$
E[S_T^2]=10600,
$$

$$
E[\ln S_T]=4.60.
$$

Thus, the derived distribution is the **least-biased probability density** consistent with the available market information and is a generalized exponential-family distribution.
